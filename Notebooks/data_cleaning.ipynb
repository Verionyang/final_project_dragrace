```python
import numpy as np
import pandas as pd

queens = pd.read_csv('../Data/raw/contestant.csv')
results = pd.read_csv('../Data/raw/episode_outcome.csv')

display(results.head())
queens.head()
```

```python
results = results.sort_values(by=['season_id','contestant','episode_num'])
results['safe']=results['outcome'].apply(lambda x: 1 if x == 'SAFE' else 0)
results['win']=results['outcome'].apply(lambda x: 1 if x == 'WIN' else 0)
results['high']=results['outcome'].apply(lambda x: 1 if x == 'HIGH' else 0)
results['low']=results['outcome'].apply(lambda x: 1 if x == 'LOW' else 0)
results['btm']=results['outcome'].apply(lambda x: 1 if x == 'BTM' else 0)

results.head(10)
```

```python
results_agg1 = results.groupby(by=['season_id','contestant'],as_index=False).agg({'safe':'sum'})
results_agg2 = results.groupby(by=['season_id','contestant'],as_index=False).agg({'win':'sum'})
results_agg3 = results.groupby(by=['season_id','contestant'],as_index=False).agg({'high':'sum'})
results_agg4 = results.groupby(by=['season_id','contestant'],as_index=False).agg({'low':'sum'})
results_agg5 = results.groupby(by=['season_id','contestant'],as_index=False).agg({'btm':'sum'})
outcome = pd.concat([results_agg1,results_agg2['win'],results_agg3['high'],results_agg4['low'],results_agg5['btm']],axis=1)
outcome.head()
```

```python
outcome.to_csv('../Data/cleaned/results.csv',index=False)
```
